```python
import math

import librosa
import numpy as np
import os
import pickle
import hmmlearn.hmm as hmm
import pyaudio
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from scipy.io.wavfile import write, read
import noisereduce as nr
from pydub import AudioSegment, effects
```

```python
def detect_leading_silence(sound, silence_threshold=-42.0, chunk_size=10):
        trim_ms = 0  # ms

        assert chunk_size > 0  # to avoid infinite loop
        while sound[trim_ms:trim_ms + chunk_size].dBFS < silence_threshold and trim_ms < len(sound):
            trim_ms += chunk_size

        return trim_ms
```

```python
def get_mfcc(file_path):
    # Trim silence
    #rate, data = read(file_path)
    #reduced_noise = nr.reduce_noise(y=data, sr=rate)
    #write("temp/trimmed.wav", rate, reduced_noise)
    
    trimmed_path = 'temp/trimmed.wav'
    sound = AudioSegment.from_file(file_path, format='wav')
    normalizedsound = effects.normalize(sound)
    normalizedsound.export(trimmed_path, format='wav')
    
    rate, data = read(trimmed_path)
    reduced_noise = nr.reduce_noise(y=data, sr=rate)
    write(trimmed_path, rate, reduced_noise)
    
    sound = AudioSegment.from_file(trimmed_path, format='wav')
    start_trim = detect_leading_silence(sound)
    end_trim = detect_leading_silence(sound.reverse())

    duration = len(sound)
    print(file_path)
    print(str(start_trim) + " - " + str(duration - end_trim))
#     if start_trim > 70:
#         start_trim -= 70
#     else: 
#         start_trim = 0
    trimmed_sound = sound[start_trim:duration - end_trim]
    #trimmed_sound = sound[:]
    trimmed_sound.export(trimmed_path, format='wav')
    
    
    y, sr = librosa.load(trimmed_path)  # read .wav file
       
    hop_length = math.floor(sr * 0.010)  # 10ms hop
    win_length = math.floor(sr * 0.025)  # 25ms frame
    # mfcc is 12 x T matrix
    mfcc = librosa.feature.mfcc(
        y=y, sr=sr, n_mfcc=12, n_fft=1024,
        hop_length=hop_length, win_length=win_length)
    # subtract mean from mfcc --> normalize mfcc
    mfcc = mfcc - np.mean(mfcc, axis=1).reshape((-1, 1))
    # delta feature 1st order and 2nd order
    delta1 = librosa.feature.delta(mfcc, order=1)
    delta2 = librosa.feature.delta(mfcc, order=2)
    # X is 36 x T
    X = np.concatenate([mfcc, delta1, delta2], axis=0)  # O^r
    # return T x 36 (transpose of X)
    return X.T  # hmmlearn use T x N matrix
```

```python
class_names = ['bat', 'tat', 'mot', 'hai', 'ba', 'sheila', 'den', 'quat']
#states = [9, 9,6,6,6, 12, 6, 9]
states = [6, 6,6,6,6, 12, 6, 6]
#dataset_path = 'dataset_s'
dataset_path = 'datasets_main_time'
model = {}
model_path = 'models_train_time'
x_train = {}
y_train = {}
x_test = {}
y_test = {}
```

```python
length = 0
for cn in class_names:
    length += len(os.listdir(f"{dataset_path}/{cn}"))
print('Total samples:', length)
```

```python
all_data = {}
all_labels = {}
for cname in class_names:
    file_paths = [os.path.join(dataset_path, cname, i) for i in os.listdir(
        os.path.join(dataset_path, cname)) if i.endswith('.wav')]
    data = [get_mfcc(file_path) for file_path in file_paths]
    all_data[cname] = data
    all_labels[cname] = [class_names.index(cname) for _ in range(len(file_paths))]
```

```python
for cname in class_names:
    x_train1, x_test1, y_train1, y_test1 = train_test_split(
        all_data[cname], all_labels[cname],
        test_size=0.2,
        random_state=1
    )
    x_train[cname] = x_train1
    x_test[cname] = x_test1
    y_test[cname] = y_test1

y_test
```

```python
total_train = 0
total_test = 0
for cname in class_names:
    train_count = len(x_train[cname])
    test_count = len(x_test[cname])
    print(cname, 'train:', train_count, '| test:', test_count)
    total_train += train_count
    total_test += test_count
print('train samples:', total_train)
print('test samples', total_test)
```

```python
for idx, cname in enumerate(class_names):
    start_prob = np.full(states[idx], 0.0)
    start_prob[0] = 1.0
    trans_matrix = np.full((states[idx], states[idx]), 0.0)
    p = 0.5
    np.fill_diagonal(trans_matrix, p)
    np.fill_diagonal(trans_matrix[0:, 1:], 1 - p)
    trans_matrix[-1, -1] = 1.0

    # trans matrix
    print(cname)
    print(trans_matrix)

    model[cname] = hmm.GaussianHMM(
        n_components=states[idx],
        verbose=True,
        n_iter=1000,
        startprob_prior=start_prob,
        transmat_prior=trans_matrix,
        params='stmc',
        init_params='stmc',
        #random_state=2,
        algorithm='viterbi'
    )
    
#     model[cname] = hmm.GMMHMM(
#         n_components=states[idx],
#         n_mix=3,
#         verbose=True,
#         n_iter=1000,
#         covariance_type='diag',
#         startprob_prior=start_prob,
#         transmat_prior=trans_matrix,
#         params='stmc',
#         init_params='st',
#         random_state=2
#     )
    
    
    model[cname].fit(X=np.vstack(x_train[cname]),
                          lengths=[x.shape[0] for x in x_train[cname]])
```

```python
for cname in class_names:
            name = f'{model_path}/model_{cname}.pkl'
            with open(name, 'wb') as file:
                pickle.dump(model[cname], file)
```

```python
print('====== Evaluation ======')
y_true = []
y_pred = []
for cname in class_names:
    for mfcc, target in zip(x_test[cname], y_test[cname]):
        scores = [model[cname].score(mfcc) for cname in class_names]
        pred = np.argmax(scores)
        y_pred.append(pred)
        y_true.append(target)
    #print( (np.array(y_true) == np.array(y_pred)).sum() / len(y_true))
print('======')
print('Confusion matrix:')
print(confusion_matrix(y_true, y_pred))
```

```python
report = classification_report( y_pred,y_true, target_names=class_names)
print(report)
```
